```python
df = pd.read_csv('repo/loan_data.csv')
df.columns
```

```python
def df_keep_outside_iqr(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)

    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(lower_bound)
    print(upper_bound)
    outside_iqr = df[(df < lower_bound) | (df > upper_bound)]
    return outside_iqr
```

### 1. person_age - Wiek Klienta

```python
df_pa_std = df_keep_outside_iqr(df['person_age'])
df_pa_std
```
