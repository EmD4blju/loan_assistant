# __Data analysis__

### Imports

```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(Path("repo", "loan_data.csv"))
df.columns
```

## To analyze
* loan_amnt
* loan_intent
* loan_int_rate
* loan_percent_income
* credit_score
* previous_loan_defaults_on_file
* loan_status

## 1. loan_amnt - amount of a loan

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['loan_amnt'], bins=30, kde=True)
plt.title('Distribution of Loan Amounts')
plt.axvline(df['loan_amnt'].mean(), color='r', linestyle='--', label='Mean')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## 2. loan_intent - purpose of a loan

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df['loan_intent'])
plt.title('Loan Intent Distribution')
plt.xlabel('Loan Intent')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## 3. loan_int_rate - interest rate

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['loan_int_rate'], bins=15, kde=True)
plt.title('Distribution of Loan Interest Rates')
plt.axvline(df['loan_int_rate'].mean(), color='r', linestyle='--', label='Mean')
plt.xlabel('Loan Interest Rate (%)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## 4. loan_percent_income - DTI (debt-to-income ratio)

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['loan_percent_income'], bins=30, kde=True)
plt.title('Distribution of DTI (Debt-to-Income Ratio)')
plt.axvline(df['loan_percent_income'].mean(), color='r', linestyle='--', label='Mean')
plt.xlabel('Loan Interest Rate (%)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# 5. credit_score - creditworthness estimate

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['credit_score'], bins=15, kde=True)
plt.title('Distribution of Credit Scores')
plt.axvline(df['credit_score'].mean(), color='r', linestyle='--', label='Mean')
plt.xlabel('Credit Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## previous_loan_defaults_on_file - any previous unpaid loans

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df['previous_loan_defaults_on_file'])
plt.title('Distribution of Previous Loan Defaults on File')
plt.xlabel('Previous Loan Defaults on File')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

## 7. loan_status - loan approval (target value)

```python
plt.figure(figsize=(10, 6))
sns.countplot(x=df['loan_status'])
plt.title('Loan Status Distribution')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
